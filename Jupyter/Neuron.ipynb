```python
import numpy as np
import matplotlib.pyplot as pl

data=np.genfromtxt('SpikesA.txt')

select=np.array([d for d in data if d[1]<50])
data1=select.transpose()

color=('red')
pl.scatter(data1[0],data1[1],alpha=0.8,s=10,c=color,edgecolors='none');
pl.title('Graph A: Spikes per neuron')
pl.xlabel('time [ms]')
pl.ylabel('Neuron ID')
pl.show();

pl.title('Rate of spikes per time')
pl.xlabel('time [ms]')
pl.ylabel('Rate [Hz]')
n,bins,patches = pl.hist(np.array(1000 * list(data1[0])), 200,normed=0, alpha=0.75)
pl.show();
```
